```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston
%matplotlib inline 
```

```python
boston_dataset = load_boston()
```

```python
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston.head()
```

```python
# Target attribute i.e. MEDV
boston_dataset.target
```

```python
# Adding the target column to the data frame.
boston['MEDV']= boston_dataset.target
```

```python
sns.set_style('whitegrid')
sns.distplot(boston['MEDV'],bins=40,kde=True)
```

From the above distplot, we can observe that MEDV is normally distributed but with some outliers.

### Correlation 
We have to select those features to fit in our linear regression model which have a high correlation with MEDV because change in them will have a greater impact on the price(MEDV)

```python
#correlation matrix
corr_matrix=boston.corr().round(3)
```

```python
#We will use heat map to plot our correlation matrix and to use the best features for our model.
plt.figure(figsize=(12, 9))
ax=sns.heatmap(corr_matrix,annot=True,cmap='inferno')

#These 2 lines fix the seaborn bug of cutting top and botton graph
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

### Observation from the heatmap 
We observe that both RM and LSTAT are correlated to MEDV 0.69 and -0.74. So we will select these 2 features in our model

```python
#Plotting these features using scatter plot
plt.figure(figsize=(25,6))

y=boston['MEDV']

plt.subplot(121)
plt.title("RM")
plt.scatter(boston['RM'],y)
plt.xlabel('MEDV')
plt.ylabel('RM')

plt.subplot(122)
plt.title("LSTAT")
plt.scatter(boston['LSTAT'],y)
plt.xlabel('MEDV')
plt.ylabel('LSTAT')
plt.show()
```

```python
X=boston[['RM','LSTAT']]
Y=boston['MEDV']
```

## Splitting data into test and train sets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2,random_state=100)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train, Y_train)
```

## Evaluation

```python
y_train_predict = model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print("\n")

# model evaluation for testing set
y_test_predict = model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
```

```python
model.score(X_test,Y_test)
```
